```python
import asyncio
import time
import aiohttp
import threading
```

```python
def get_data_from_db(query: str):
    print(f'Получен запрос: {query}')
    if query == "Краснодар":
        time.sleep(4)
    elif query == "Адлер":
        time.sleep(2)
    print(f'Данные получены {query}')
```

```python
async def get_data_from_db_async(query: str):
    print(f'Получен запрос: {query}')
    if query == "Краснодар":
        await asyncio.sleep(4)
    elif query == "Адлер":
        await asyncio.sleep(2)
    print(f'Данные получены {query}')

    return "result"
```

```python
get_data_from_db('Краснодар')
get_data_from_db('Адлер')
```

```python
await get_data_from_db_async('Краснодар')
await get_data_from_db_async('Адлер')
```

```python
await asyncio.gather(
    get_data_from_db_async('Краснодар'),
    get_data_from_db_async('Адлер'),
    get_data_from_db_async('Краснодар'),
    get_data_from_db_async('Адлер'),
    get_data_from_db_async('Краснодар'),
    get_data_from_db_async('Адлер'),
    get_data_from_db_async('Краснодар'),
    get_data_from_db_async('Адлер'), 
)
```

```python
task1 = asyncio.create_task(get_data_from_db_async('Краснодар'))
task2 = asyncio.create_task(get_data_from_db_async('Адлер'))
```

```python
async def get_data(i: int, endpoint: str):
    url = f'http://127.0.0.1:8000/{endpoint}/{i}'
    print(f'Начато выполнение: {i}')
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as responce:
            print(f'Закончено выполнение: {i}')
```

```python
await get_data(1, 'sync')
```

```python
await asyncio.gather(
    *[get_data(i, 'sync') for i in range(300)]
)
```

```python
await asyncio.gather(
    *[get_data(i, 'async') for i in range(300)]
)
```
